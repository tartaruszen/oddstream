# Oddstream - Oil Pipeline Example

```python
%matplotlib inline
from matplotlib.pylab import rcParams
import matplotlib.pyplot as plt
rcParams['figure.figsize'] = 15, 6

from oddstream.oddstream import OddStreams
from sklearn.preprocessing import MinMaxScaler
from sklearn import preprocessing
import numpy as np
import pandas as pd
```

## Create Training Data

### Parameters

```python
config = {'nIn': 100, 'dropout_rate': 0.1, 'epochs': 500, 'batch_size': 256, 'validation_split': .1}
max_iterations = 100
trials = 100
```

### Load & Prepare Data

```python
anomalous_stream = pd.read_csv("anomalous_stream.csv")
anomalous_stream = np.array(anomalous_stream)
n = anomalous_stream.shape[1]
X = np.array([])
for i in range(0, n):
    X = np.append(X, anomalous_stream[0:config['nIn'], i])
X = X.reshape(n, config['nIn'])  
```

```python
plt.plot(anomalous_stream[:, :], alpha=0.01, color="blue")
plt.show
```

### Scale Data

```python
scaler = MinMaxScaler()
scaler.fit(X=X)
X = scaler.transform(X).reshape(n, config['nIn'], 1)
```

## Initialization and Training

```python
oddstream = OddStreams(max_iterations = max_iterations, trials = trials, config = config, useCUDNN = True)
```

```python
oddstream.train(X)
```

## Simulate Streaming

```python
n_steps = int((anomalous_stream.shape[0] - config['nIn']) / config['nIn'])
outlier = {}
for j in range(0, n_steps):
    # create input
    X_test = np.array([])
    for i in range(0, n):
        X_test = np.append(X_test, anomalous_stream[(j + 1) * config['nIn']:((j + 2) * config['nIn']), i])
    X_test = X_test.reshape(n, config['nIn'])
    X_test = scaler.transform(X_test).reshape(n, config['nIn'], 1)
    # predict outliers
    outlier[j] = oddstream.predict(X_test)
```

```python
outlier_matrix = np.array([[0.0 for i in range((n_steps + 1) * config['nIn'])] for j in range(n)], dtype=float)
for i in range(0, n_steps):
    outlier_matrix[outlier[i], (i + 1) * config['nIn']:((i + 2) * config['nIn'])] = 1.0
```

```python
plt.imshow(outlier_matrix, cmap='hot', interpolation='nearest')
plt.show()
```
